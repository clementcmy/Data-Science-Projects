## Background

With an ageing population, the cost of healthcare which Singapore needs to pay is projected to rise. As such, the government had come up with the HealthierSG initiative. Healthier SG is a national initiative which focuses mainly on preventive health (i.e., steps to lead healthier lifestyle in order to prevent the onset on chronic diseases). 

This purpose of this notebook is to conduct a simple EDA to analyse the trends of visits made to Singapore Healthcare institutions (e.g., Hospitals, Polyclinics etc.) from historical data.

## Data Cleaning & Inspection

### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Inspection

```python
df = pd.read_csv("./Data/Healthcare Admissions Data.csv")
```

```python
df.head(10)
```

Unnecessary rows from the dataset will be removed. 

```python
df.drop(range(0,9),inplace=True)
df.head()
```

```python
df.drop(range(23,69),inplace = True)
df
```

The next few steps below will involve inspecting and renamine the rows and columns.

```python
# Rename header using first row
df.rename(columns=df.iloc[0], inplace = True) 

# Drop first row
df.drop(range(9,10), inplace=True)

# Reset index
df.reset_index(drop=True, inplace=True) 
df
```

```python
# Rename index
df.rename(index={0:"Acute Hospitals Admissions (AHA)", 1:"AHA Public", 2:"AHA Non-Public",
                 3:"Psychiatric Hospitals Admissions (PHA)", 4:"PHA Public", 5:"PHA Non-Public",
                 6:"Community Hospitals Admissions (CHA)", 7:"CHA Public", 8:"CHA Non-Public",
                 9:"Specialist Outpatient Clinics", 10:"Accident & Emergency Departments",
                 11:"Polyclinics", 12:"Dental Clinics"}, inplace=True)
# Drop first column
df.drop(columns="Data Series", inplace=True)

# Check index names
df
```

```python
# Transpose columns to rows
df = df.T 
df
```

```python
# Change index to datetime
df.index = pd.to_datetime(df.index)
df
```

From the above table, we can see that there are na in some of the cells. Hence, the next steps will address the following.
<ul>
    <li>Check for duplicates</li>
    <li>Check for null values and replace them with NaN</li>
    <li>Check data types of the dataset </li>
</ul>

```python
# Check for duplicate data
df.duplicated().sum()
```

No duplicated data was observed.

```python
# Check for null values
df.isnull().sum()
```

There are no null values in the cells. Therefore, the 'na' will be replaced with NaN in the step below.

```python
# Replace na with null
df = df.replace('na', np.nan)
df.tail()
```

```python
# Check data types of the dataset
df.info()
```

After inspecting the data type, we can see that they are all object. We will need to convert the data type to numbers for us to analyse it. Hence, the data type will be converted to float.

```python
# Change data type to float
df = df.astype(float)
```

```python
df.info()
```

## Data Visualisation

To have an overall idea on the maximum no. of visitors to each type of healthcare intuitions, the values will be displayed below.

```python
for i in df:
    print(f'{i}: {df[i].max()}')
```

### Figure 1: Overview of visits across the various types of healthcare intuitions

```python
plt.figure(figsize=(7, 10))
plt.subplot(7,1,1)
plt.plot(df.index,df['Acute Hospitals Admissions (AHA)'], color='purple')
plt.title('Acute Hospitals Admissions (AHA)')

plt.subplot(7,1,2)
plt.plot(df.index,df['Psychiatric Hospitals Admissions (PHA)'], color='orange')
plt.title('Psychiatric Hospitals Admissions (PHA)')

plt.subplot(7,1,3)
plt.plot(df.index,df['Community Hospitals Admissions (CHA)'], color='green')
plt.title('Community Hospitals Admissions (CHA)')

plt.subplot(7,1,4)
plt.plot(df.index,df['Specialist Outpatient Clinics'], color='fuchsia')
plt.title('Specialist Outpatient Clinics')

plt.subplot(7,1,5)
plt.plot(df.index,df['Accident & Emergency Departments'], color='red')
plt.title('Accident & Emergency Departments')

plt.subplot(7,1,6)
plt.plot(df.index,df['Polyclinics'], color='deepskyblue')
plt.title('Polyclinics')

plt.subplot(7,1,7)
plt.plot(df.index,df['Dental Clinics'], color='olive')
plt.title('Dental Clinics')

plt.tight_layout()
```

In general, it seems that the no. of visitors across healthcare intuitions are steadily increasing over the years. However, it is interesting to note that visits to dental clinics have somewhat decrease over the years. In addition, if we narrow down and look at Psychiatric Hospital Admissions, it is worrying to see that the no. of visits have increased sharply from 2022 to 2023, stressing the importance of mental health in Singapore.

### Figure 2: Overview of public visits to Acute, Psychiatric and Community Hospitals

```python
plt.subplot(3,1,1)
plt.plot(df.index,df['AHA Public'], color='purple')
plt.title('AHA Public')

plt.subplot(3,1,2)
plt.plot(df.index,df['PHA Public'], color='orange')
plt.title('PHA Public')

plt.subplot(3,1,3)
plt.plot(df.index,df['CHA Public'], color='green')
plt.title('CHA Public')

plt.tight_layout()
```

Similar to Figure 1, the amount of public visits to the various types of hospitals have been increasing over the years.

### Figure 3: Overview of non-public visits to Acute, Psychiatric and Community Hospitals

```python
plt.subplot(3,1,1)
plt.plot(df.index,df['AHA Non-Public'], color='purple')
plt.title('AHA Non-Public')

plt.subplot(3,1,2)
plt.plot(df.index,df['PHA Non-Public'], color='orange')
plt.title('PHA Non-Public')

plt.subplot(3,1,3)
plt.plot(df.index,df['CHA Non-Public'], color='green')
plt.title('CHA Non-Public')
plt.tight_layout()
```

Unlike Figure 1 and 2, it is interesting that the non-public visits to the various types of hospitals have decreased. With non-public psychiatric visits having no visits since 2008. Could this mean that Singapore's healthcare revenue from private sources have been decreasing?
